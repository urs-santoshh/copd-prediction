# COPD in Nepal

### Factors that lead to COPD
1. Biomass Fuel Exposure (Indore air pollution)
2. Tobacco smoking
3. Outdoor air pollution (Vehicle emission, industrial smoke, dust form constructions etc.)
4. Occupational exposure (People working in agriculture, manufacruring industries, construction etc.)
5. Genetic susceptibility
6. Respiratory infections in the childhood
7. Low awareness and health access
9. Climate and environmental factors

### Signs and symptoms of COPD
1. Chronic cough
2. Shortness of breath (dyspnea)
3. Excess mucous productions (Sputum) (Phlegm)
4. Wheezing (high-pitched whistling sound made while breathing)
5. Fatigue and low energy
6. Chest tightness
7. Frequent resipratory infections
8. Cynosis (medical term for when your skin, lips or nails turn blue due to a lack of oxygen in your blood)
9. Unintended weight loss
10. Chronic bronchitis and emphysema symptoms
11. Frequent Falre-ups (Exacerbations)

### Univariate Analysis

#### i. Numerical features

```python
# what is the age distribution in my data
sns.histplot(data=data["Age"], bins=10, kde=True, color="blue")
plt.title("Age distribution")
plt.show()
```

```python
# BMI distribution
sns.histplot(data=data["BMI"], bins=10, kde=True, color="Green")
plt.title("BMI distribution")
plt.show()
```

```python
# Duration of smoking distribution
sns.histplot(data=data["Duration_Of_Smoking"], bins=10, kde=True, color="Red")
plt.title("Duration of Smoking distribution")
plt.show()
```

#### ii. Categorical features

```python
# gender distrubution
sns.countplot(x = "Gender", data=data, palette="viridis")
plt.title("Gender distribution")
plt.show()
```

```python
# smoking status
sns.countplot(x = "Smoking_Status", data=data, palette="viridis")
plt.title("Smoking status")
plt.show()
```

### Bivariate anlaysis

```python
plt.figure(figsize=(14,10))
```

```python
# Age vs COPD diagnosis
sns.boxenplot(x = "COPD_Diagnosis",y="Age", data=data, palette="coolwarm")
plt.title("Age vs COPD diagnosis")
plt.show()
```

```python
# BMI vs COPD diagnosis
sns.boxenplot(x = "COPD_Diagnosis", y = "BMI", data = data, palette="viridis")
plt.title("BMI vs COPD diagnosis")
plt.show()
```

### Multivariate analysis

```python
# pairplot
sns.pairplot(data = data, hue="COPD_Diagnosis", diag_kind="kde", palette="Set1")
plt.title("Pair plot")
plt.show()
```

```python
# correlation matrix
data_corr = data[["Age","Duration_Of_Smoking", "Biomass_Fuel_Exposure", "Occupational_Exposure", "Family_History_COPD", "BMI", "Air_Pollution_Level","Respiratory_Infections_Childhood", "COPD_Diagnosis"]]
corr = data_corr.corr()
```
